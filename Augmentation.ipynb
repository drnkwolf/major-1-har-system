```python
import numpy as np
import pandas as pd
import csv
```

```python
source = 'D:/UPES/UPES SEM VII/Major 1/endsem/HAPT Data Set/SpectroSet/200feature/'
dest = 'D:/UPES/UPES SEM VII/Major 1/endsem/HAPT Data Set/AugmentedData/200feature/'
fs = 50
```

```python
for i in range(1,13):
#     print(i)
    file = source + str(i) + '_SpectroFeatures.csv'
    df = pd.read_csv(file,header=None)
#     print(df.shape)
    x = df.rolling(4).mean()
    x.dropna(inplace=True)
#     print(x.shape)
    df = df.append(x)
#     print(df.shape)
    df = df.sample(frac=1).reset_index(drop=True)
    x = df.rolling(2).mean()
    x.dropna(inplace=True)
    df = df.append(x)
    df = df.reset_index(drop=True)

#     print(df.shape)

    with open (dest+str(i)+'_AugSpectroFeatures.csv','a',newline='') as out_file:
        writer = csv.writer(out_file)
        for j in range(df.shape[0]):
            k = df.ix[j]
            writer.writerow(np.concatenate((k,[i])))
```
